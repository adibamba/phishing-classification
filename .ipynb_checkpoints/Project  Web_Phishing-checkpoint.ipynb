# 1) DESCRIPTION OF THE DATASET

This dataset is designed for the classification task of identifying whether a given website is legitimate or potentially phishing. It contains a total of 1,353 website instances, each characterized by 9 numerical features and a binary target variable indicating the classification result.

Dataset Characteristics

- Dataset: Website Phishing
- Domain: Cybersecurity / Web Safety
- Type: Classification (Binary)
- Total Instances: 1,353
- Number of Features: 9 predictive features + 1 target variable
- Target Variable: Result â€” Indicates whether the website is legitimate (0) or phishing (1)
- Missing Values: None
